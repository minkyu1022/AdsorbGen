# IS2RES false-positive anomaly inspection

30 random samples that **we flagged as anomalous** but **OC20 official label says clean**.
Each viewer shows init → final as a 2-frame trajectory; use the slider to flip.
Filename encodes which check fired and max displacement in Å.

```python
import re
from pathlib import Path
from ase.io import read
from ase.visualize import view

DIR = Path('../../runs/probe_v3/fp_inspect').resolve()
assert DIR.exists(), f'not found: {DIR}'

pairs = {}
for f in sorted(DIR.glob('fp_*.xyz')):
    m = re.match(r'(fp_sid\d+_idx\d+_.+?_dmax[\d.]+)_(init|final)\.xyz', f.name)
    if not m:
        continue
    base, kind = m.group(1), m.group(2)
    pairs.setdefault(base, {})[kind] = f

keys = sorted(pairs)
print(f'dir: {DIR}')
print(f'{len(keys)} sample pairs')
for k in keys:
    print(' ', k)
```

```python
def show(i):
    base = keys[i]
    p = pairs[base]
    init = read(p['init'])
    final = read(p['final'])
    init.center()
    final.center()
    print(f'[{i}] {base}')
    print(f'  N_atoms={len(init)}  cell_abc={init.cell.cellpar()[:3].round(2)}')
    print('  init:')
    display(view(init, viewer='x3d'))
    print('  final:')
    return view(final, viewer='x3d')
```
